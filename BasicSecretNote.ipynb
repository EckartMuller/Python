```python
import os
class SecretNote():
    def __init__(self):
        print("Secret Note")
    def notOlustur(self, notName, notSifre):
        self.__notName = notName
        self.notSifre = notSifre
        
        if os.path.exists(self.__notName+".txt"):
            notSilSilme = input("Zaten boyle bir not var silinsin mi ?(Y/N)")
            if notSilSilme == "Y":
                os.remove(self.__notName+".txt")
                print("Not silindi.!")
            
        else:
            f = open(self.__notName+".txt", "x")
    def notSifreKoy(self, notSifre):
        self.__notSifre = notSifre

    
    def notIcerıkEkle(self, notIcerık):
        try:
            with open(self.__notName+".txt", mode="w") as notum:
                notum.write(notIcerık)
        except:
            print("Nota yazı yazılamadı")

    
    def notSil(self):
        if os.path.exists(self.__notName+".txt"):
            os.remove(self.__notName+".txt")
        else:
            print("Boyle bir not zaten yok")

    
    def notOku(self, notOkuSifre):
        self.__notOkuSifre = notOkuSifre
        if self.__notSifre == self.__notOkuSifre:
            try:
                with open(self.__notName+".txt", mode="r") as notum:
                    notIcerik = notum.read()
                    print(notIcerik)
            except:
                print("Hata")
        else:
            print("Sifre dogru değil")
```
